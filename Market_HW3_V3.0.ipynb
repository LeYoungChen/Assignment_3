```python
# Calculate the total sales and marketing cost based on 3 tiers

tier1 = {'email': 1000.0, 'social': 1000.0000000000001, 'display': 1000.0, 'paid_search': 999.9999999999999, 'referral': 1000.0, 'organic_search': 0.0, 'direct': 0.0}
tier2 = {'email': 2000.0, 'social': 2000.0, 'display': 2000.0, 'paid_search': 2000.0, 'referral': 2000.0, 'organic_search': 0.0, 'direct': 0.0}
tier3 = {'email': 3000.0, 'social': 3000.0, 'display': 3000.0, 'paid_search': 3000.0, 'referral': 3000.0, 'organic_search': 0.0, 'direct': 0.0}
tier_all = [tier1,tier2,tier3]

def dict_add(x,y):        
    for k, v in y.items():
        if k in x.keys():
            x[k] += v
        else:
            x[k] = v
    return(x)
```

```python
tier_sum = dict_add(tier1, tier2)
tier_sum = dict_add(tier_sum, tier3)
```

```python
data = data[data['convert_TF'] == True]
data
```

```python
data= data.drop(['convert_TF','tier'],axis = 1)
```

The three attribution methods are last interaction, first interaction, last non-direct interaction

## Last interaction

```python
last_list = data.values.tolist()
last_list
```

```python
last_list_1=[]
for i in last_list:
    last_list_1.append([x for x in i if x == x])
last_list_1
```

```python
# This is list is without NaN
```

```python
last_list_final = []
for i in last_list_1:
    last_list_final.append(i[-1])
last_list_final
```

## First interaction

```python
first_list_final = []
for i in last_list_1:
    first_list_final.append(i[0])
```

## Last non-direct interaction

```python
nondirect_list_1=[]
for i in last_list_1:
    nondirect_list_1.append([x for x in i if x == x and x!='direct'])
nondirect_list_1
```

```python
nondirect_list_final = []
for i in nondirect_list_1:
    if len(i) > 0:
        nondirect_list_final.append(i[-1])
    else:
        break
```

## Average CAC

```python
# Here we calculate the cac for last interaction

cac_last = {}
for key in tier_sum.keys():
    cac_last[key] = int(tier_sum[key])/int(last_list_final.count(key))
cac_last
```

```python
# Here we calculate the cac for first interaction

cac_first = {}
for key in tier_sum.keys():
    cac_first[key] = int(tier_sum[key])/int(first_list_final.count(key))
cac_first
```

```python
# Here we calculate the cac for last non-direct interaction

cac_last_non = {}
for key in tier_sum.keys():
    if key != 'direct':
        cac_last_non[key] = int(tier_sum[key])/int(nondirect_list_final.count(key))
    if key == 'direct':
        cac_last_non[key] = 0
cac_last_non
```

```python
data2 = pd.read_csv("attribution_allocation_student_data.csv")
```

```python
tier1 = {'email': 1000.0, 'social': 1000.0000000000001, 'display': 1000.0, 'paid_search': 999.9999999999999, 'referral': 1000.0, 'organic_search': 0.0, 'direct': 0.0}
tier2 = {'email': 2000.0, 'social': 2000.0, 'display': 2000.0, 'paid_search': 2000.0, 'referral': 2000.0, 'organic_search': 0.0, 'direct': 0.0}
tier3 = {'email': 3000.0, 'social': 3000.0, 'display': 3000.0, 'paid_search': 3000.0, 'referral': 3000.0, 'organic_search': 0.0, 'direct': 0.0}
```

```python
data2 = data2[data2['convert_TF'] == True]
```

```python
data_t1 = data2[data2['tier'] == 1]
data_t2 = data2[data2['tier'] == 2]
data_t3 = data2[data2['tier'] == 3]
```

```python
data_t1 = data_t1.drop(['convert_TF','tier'],axis = 1)
data_t2 = data_t2.drop(['convert_TF','tier'],axis = 1)
data_t3 = data_t3.drop(['convert_TF','tier'],axis = 1)
```

```python
t1_list = data_t1.values.tolist()
t2_list = data_t2.values.tolist()
t3_list = data_t3.values.tolist()
```

```python
first_list_1 = []
for i in t1_list:
    first_list_1.append(i[0])
    
first_list_2 = []
for i in t2_list:
    first_list_2.append(i[0])
    
first_list_3 = []
for i in t3_list:
    first_list_3.append(i[0])
```

```python
margin_cac_t1 = {}
for key in tier1.keys():
    margin_cac_t1[key] = int(tier1[key])/int(first_list_1.count(key))
margin_cac_t1
```

```python
margin_cac_t2 = {}
for key in tier2.keys():
    if margin_cac_t1[key] > 0:
        margin_cac_t2[key] =  (tier2[key]-tier1[key]) / ((tier2[key])/(int(tier2[key])/int(first_list_2.count(key))) - (tier1[key])/margin_cac_t1[key]) 
margin_cac_t2
```

```python
margin_cac_t3 = {}
for key in margin_cac_t2.keys():
    margin_cac_t3[key] =  (tier3[key]-tier2[key]) / ((tier3[key])/(int(tier3[key])/int(first_list_3.count(key))) - (tier2[key])/margin_cac_t2[key]) 
margin_cac_t3
```

```python
margin_cac_t1_list = [margin_cac_t1]
margin_cac_data = pd.DataFrame(margin_cac_t1_list)
```

```python
margin_cac_data = margin_cac_data.append(margin_cac_t2, ignore_index=True)
margin_cac_data = margin_cac_data.append(margin_cac_t3, ignore_index=True)
```
